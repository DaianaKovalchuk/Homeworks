## Task 1 (0.5 point)

Write a program that receives a positive integer `N` from the user, the task is to find following formula:

$$
1^2 + 2^3 + 3^2 + ... + N^2 (or N^3)
$$

1. If the number is odd then rais to the power of 2.
2. If the number is even then rais to the power of 3.


Example:

$$
N = 5
$$

$$
1^2 + 2^3 + 3^2 + 4^3 + 5^2 = (1 + 8 + 9 + 64 + 25) = 107
$$

The program should work for any arbitrary N from the user.

````python
n = int(input('Type your number: '))
numbers = range(0,n+1)
sum=0
for i in numbers:
    if i%2==0:
      sum+=i**3
    else:
      sum+=i**2
print(sum)
````

## Task 11 - Shifting 2D Matrix (1 point)

Let's develop the idea from the task in previous homework, now consider a 2D array, and you want to implement a program that not only shift to right or left, but also, it can shift up or down.

In this scenario, the user need to insert four different arguments:

1. Row shift or column shift? ('row' or 'column')

2. Which direction? ('left', 'right' or 'top', 'down')

3. Index of the column or row? (an integer index)

4. How many elements need to be shifted? (integer number)

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
q1 = input('row or column? ')
q2 = input('direction? (left, right, top or down) ')
q3 = int(input('index? '))
q4 = int(input('how many elements need to be shifted? '))
if q1 == 'row':
    row = matrix[q3]
    if q2 == 'left':
        row = row[q4:] + row[:q4]
    else:
        row = row[-q4:] + row[:-q4]
    matrix[q3] = row
else:
    column = []
    for i in range(len(matrix)):
        column.append(matrix[i][q3])
    if q2 == 'top':
        column = column[q4:] + column[:q4]
    else:
        column = column[-q4:] + column[:-q4]
    for i in range(len(matrix)):
        matrix[i][q3] = column[i]
print("\nResult:")
for row in matrix:
    print(row)
````

## Task 12 - Validating Password (1 points)

A website requires the users to input username and password to register. Write a program to check the validity of password input by users.

Following are the criteria for checking the password:

1. At least 1 letter between [a-z]
2. At least 1 number between [0-9]
1. At least 1 letter between [A-Z]
3. At least 1 character from \[$#@]
4. Minimum length of transaction password: 6
5. Maximum length of transaction password: 12

Your program should accept a sequence of comma separated passwords and will check them according to the above criteria. Passwords that match the criteria are to be printed, each separated by a comma.

```
If the following passwords are given as input to the program:
Input: ABd1234@1,a F1#,2w3E*,2We3345

Then, the output of the program should be:
Output: ABd1234@1
```

Hint: You may check the function **re.search()** function (regular expression).

````python
password = input('Your password: ')
check = password.split(',')
v_password = []
for word in check:
  word = word.strip()
  if 6<=len(word)<=12:
    is_lower = False
    is_upper = False
    is_number = False
    is_symbol = False
    for letter in word:
      if 'a'<= letter <= 'z':
        is_lower = True
      elif 'A' <= letter <= 'Z':
        is_upper = True
      elif '0' <= letter <= '9':
        is_number = True
      elif letter == '$' or letter == '@' or letter == '#':
        is_symbol = True
    if is_lower and is_upper and is_number and is_symbol:
      v_password.append(word)
print(','.join(v_password))
````

````python
import re
password = input('Your password: ')
check = password.split(',')
v_password = []
pattern = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[$#@])[A-Za-z\d$#@]{6,12}$'
for word in check:
  word = word.strip()
  if re.search(pattern, word):
    v_password.append(word)
print(','.join(v_password))
````
